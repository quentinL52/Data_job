```python
from transformers import AutoTokenizer, AutoModel

# Sélectionner les colonnes de metadata
metadata = ['entreprise', 'publication', 'poste', 'contrat']

# Combiner les colonnes de metadata avec la description
df['combined_text'] = df.apply(lambda row: ' '.join([str(row[col]) for col in metadata]) + ' ' + row['description'], axis=1)

# Charger le modèle et le tokenizer
tokenizer = AutoTokenizer.from_pretrained('distilbert-base-uncased')
model = AutoModel.from_pretrained('distilbert-base-uncased')

# Fonction pour obtenir l'embedding
def get_embedding(text):
    inputs = tokenizer(text, return_tensors='pt', max_length=512, truncation=True, padding=True)
    outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).detach().numpy()

# Appliquer la fonction d'embedding
df['embedding'] = df['combined_text'].apply(lambda x: get_embedding(x))

# Afficher le dataframe avec les embeddings
print(df[['id', 'combined_text', 'embedding']])
```

```python
import fitz  # PyMuPDF

def extract_text_from_pdf(pdf_path):
    # Ouvrir le document PDF
    document = fitz.open(pdf_path)
    text = ""

    # Parcourir les pages du document
    for page_num in range(len(document)):
        page = document.load_page(page_num)
        text += page.get_text()

    return text

# Spécifiez le chemin vers votre fichier PDF
pdf_path = '/content/drive/MyDrive/wild code school/CV Quentin.pdf'
cv_text = extract_text_from_pdf(pdf_path)
print(cv_text)
```

# methode creation d'un index en supprimant le precedant pour le cv

```python
def cv_index(documents):
  pc = Pinecone(api_key=pinecone_api_key)
  index_name = "cv"
  pc.delete_index(index_name)
  existing_indexes = [index_info["name"] for index_info in pc.list_indexes()]

  if index_name not in existing_indexes:
      pc.create_index(
          name=index_name,
          dimension=768, # fonctionne pour 768 dimensions mais le probleme est que le modele met enormement de temps a charger
          metric="cosine",
          spec=ServerlessSpec(cloud="aws", region="us-east-1"),
      )
      while not pc.describe_index(index_name).status["ready"]:
          time.sleep(1)

  index = pc.Index(index_name)
  docsearch = PineconeVectorStore.from_documents(documents, embeddings, index_name=index_name,pinecone_api_key=pc)
  return docsearch
```
