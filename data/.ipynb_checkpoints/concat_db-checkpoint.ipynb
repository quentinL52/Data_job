```python
import sqlite3
import pandas as pd
import numpy as np

def connect_db(file, table):
    # Connectez-vous à la base de données
    conn = sqlite3.connect(f"./{file}.db")
    
    # Supposez que vous voulez lire une table nommée 'nom_table'
    table_name = table
    df = pd.read_sql_query(f"SELECT * FROM {table_name}", conn)
    conn.close()
    return df
```

```python
df_apec = connect_db("apec","df_clean")
```

```python
df_wtj = connect_db("welcometothejungle","df_clean")
```

```python
df = pd.concat([df_apec, df_wtj]).reset_index(drop=True)
doublons = df[df.duplicated(subset=['id'], keep=False)]
```
